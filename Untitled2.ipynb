```python
# Loading the packages
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pandas import read_csv
df= pd.read_csv (r'/content/sales_data.csv')
```

```python
df.head(8)
```

```python
df.describe()
```

```python


df.fillna(df.mean(),inplace=True)
```

```python
df
```

```python


df['revenue'] = df['Quantity'] * df['Price']
```

```python


df.plot(x='Date', y='revenue')
plt.title('Revenue vs Date')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()
```

```python


print('Total orders:', df['Quantity'].sum())
print('Total revenue:', df['revenue'].sum())
```

```python


df.groupby('Product')['Price'].mean()
```

```python
# prompt: make the above as a table

df.groupby('Product')['Price'].mean().to_csv('mean_price.csv')
```

```python


df.groupby('Product')['Price'].mean().plot(kind='bar')
plt.title('Average Price by Product')
plt.xlabel('Product')
plt.ylabel('Average Price')
plt.show()
```

```python


max_value = max(df.groupby('Product')['Price'].mean())
max_product = df.groupby('Product')['Price'].mean().idxmax()
print('Max value:', max_value)
print('Max product:', max_product)
```
